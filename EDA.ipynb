**Ejercicio 1:** Por pais muestre cual fue el año en el que mejor ingreso se registro y su valor, adicional muestre cual fue el año con mas Invoices (Frecuencia) y otra columna con el total.

Nota: Note que el año con mejor ingreso no es lo mismo con el año con mas invoices, por ejemplo en 2019 Honduras en dos invoices se facturo 10,000. sin embargo, en 2020 se realizaron 20 invoices (pueda que en este año el ingreso de esas invoices hayan sumado 7,000 inferior al 2020) 

*recuerde que el ejemplo es inventado*

Ejemplo de como se deberia de ver el dataframe resultante:


```python
pd.DataFrame({
    "country" : ["Honduras"]
    , "YearMejorIngreso" : [2019]
    , "ValorYearMejorIngreso" : [20000]
    , "YearMejorFrecuenciaInvoices": [2020]
    , "CantidadYearMejorFrecuenciaInvoices": [20]
})
```

*Answer:*

```python
ecommerce1 = ecommerce.sort_values(['Country', 'year'] , ascending=True)
```

```python
ecommerce1['InvoiceTotal'] = (ecommerce1['Quantity'] * ecommerce1['UnitPrice'])
```

Creando sub tablas con los datos deseados

```python
mTotal = ecommerce1.groupby(['Country', 'year'], as_index=False)['InvoiceTotal'].sum()
```

```python
mTotal = mTotal.groupby(['Country'],  as_index=False).max()
```

```python
mInvoice = ecommerce1.groupby(['Country', 'year'], as_index=False)['InvoiceNo'].count()
```

```python
mInvoice = mInvoice.groupby(['Country'],  as_index=False).max()
```

Uniendo las tablas encontradas

```python
results = mTotal.merge(mInvoice, on=["Country"], how = "inner")
```

```python
results = results.rename(columns = {
    'year_x' : 'YearMejorIngreso', 
    'InvoiceTotal': 'ValorYearMejorIngreso', 
    'year_y': 'YearMejorFrecuenciaInvoices', 
    'InvoiceNo' : 'CantidadYearMejorFrecuenciaInvoices'})
```

```python
results
```

Ejemplo de como se deberia de ver el dataframe resultante (Recuerde que este es un ejemplo inventado): 

```python
pd.DataFrame({
    "Country" : ["Canada","Canada","Canada","Spain","Spain","Spain"]
    , "Top" : [1,2,3,1,2,3]    
    , "StockCode": [546,6886,665,656,888,668]
    , "Description": ["Product X","Product K","Product J","Product C","Product D","Product U"]
    , "TotalOnInvoices": [2000,2100,4000,3000,3800,3900]
})
```

*Answer:*

Ordenando y sacando total de facturacion

```python
ecommerce4 = ecommerce.sort_values(['Country', 'StockCode'] , ascending=True)
```

```python
ecommerce4['TotalOnInvoices'] = (ecommerce['Quantity'] * ecommerce['UnitPrice'])
```

Tomando el ultimo año

```python
year = ecommerce4['year'].max()
```

```python
ecommerce4 = ecommerce4.loc[ecommerce4['year'] == year, :]
```

Sumando los totales para cada producto en el año

```python
ecommerce4 = ecommerce4.groupby(['Country', 'StockCode', 'Description'], as_index=False)['TotalOnInvoices'].sum()
```

Ordenando de menos vendido al mas vendido

```python
ecommerce4 = ecommerce4.sort_values(['Country', 'TotalOnInvoices'] , ascending=True)
```

Tomando el top 3 menos facturados

```python
results = ecommerce4.groupby(['Country'], as_index=False).head(3)
```
